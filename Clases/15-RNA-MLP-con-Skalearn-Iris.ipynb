## Perceptrón Multicapa (MLP)

<li>Es uno de los tipos mas fundamentales</li>
<li>Consiste en mlitiples capas de neuronas conectas entre si</li>
<li>Se utiliza para tareas de clasificacion y regresion</li>

```python
# cargar paquetes
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
# cargar datos 
data = sns.load_dataset('iris')
data.head()
```

```python
# crear datos de prueba y de entrenamiento
train, test = train_test_split(data, test_size=0.4, stratify=data['species'] ,random_state=31)
```

```python
# variables de entrenamiento
x_train = train[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_train = train['species']

# variables de prueba
x_test = test[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y_test = test['species']
```

```python
# crear modelo 
mlp = MLPClassifier(hidden_layer_sizes=(3, 3), solver='sgd', learning_rate_init=0.01, max_iter=500)
```

```python
data['species'].unique()
```

```python
# codificar la variable de salida(de repuesta), porque  es categórica
from sklearn.preprocessing import LabelEncoder
codificador = LabelEncoder()
```

```python
data['species'] = codificador.fit_transform(data['species'])
species = pd.DataFrame({'species': ['setosa', 'versicolor', 'virginica']})
species
```

```python
#data.iloc[70:80]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
# guardar datos no escalados 
noescalado = x_train.copy()
x_train_array = sc.fit_transform(x_train.values)
```

```python
# guardar datos escalados
x_train = pd.DataFrame(x_train_array, index=x_train.index ,columns=x_train.columns)
```

```python
x_test_array = sc.fit_transform(x_test.values)
x_test = pd.DataFrame(x_test_array, index=x_test.index ,columns=x_test.columns)
x_test
```

```python
# entrenar el modelo
mlp.fit(x_train, y_train)
```

```python
# ver lo que paso, la experiencia de la red
mlp.coefs_
```

```python
# aplicar la predicción

from sklearn.metrics import accuracy_score 
```

```python
# ver la eficienca del modelo 
mlp.score(x_train, y_train)
```

```python
prediccion_train = mlp.predict(x_train)
prediccion_test = mlp.predict(x_test)
```

```python
train_puntero = accuracy_score(prediccion_train, y_train)
test_punteo = accuracy_score(prediccion_test, y_test)
print("Puntaje de entrenamiento: ", train_puntero*100,"%")
print("Puntaje de prueba: ", test_punteo*100,"%")
```
