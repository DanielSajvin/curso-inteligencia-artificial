## Web Scraping

Se requiere el módulo BeautifulSoup

```python
import requests 
from bs4 import BeautifulSoup
```

```python
# enlace a la pagina de consulta 
url = 'https://es.wikipedia.org/wiki/Organizaci%C3%B3n_territorial_de_Guatemala'
```

```python
response = requests.get(url)
```

```python
print('Status Code', url, ':', response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser') # se convierte la respuesta a texto plano
```

```python
tabla = soup.find('table', {'class': 'wikitable'})
```

```python
# dir(tabla)
import pandas as pd
```

```python
tabla = pd.read_html(str(tabla))
tabla[1]
```

```python
departamento = tabla[1]
```

```python
departamento
```

```python
departamento.columns = ['Departamento', 'Cabecera', 'Superficie', 'Población']
departamento
```

```python
departamento.describe()
```
